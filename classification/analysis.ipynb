```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

### Plot SVM and PCA line

```python
pvc200=[0.9529,0.9547,0.9525]
pvc400=[0.9647,0.9625,0.9623]
pvc600=[0.9621,0.9599, 0.9597]
pvc800=[0.9598,0.9599, 0.9574]
pvc1000=[0.9598, 0.9599, 0.9574]
pvc1200=[0.9598, 0.9599, 0.9574]
nopvc=[0.9598,0.9599, 0.9574]
tit=[200,400,600,800,1000,1200]
pvcpre=[0.9529,0.9647,0.9621,0.9598,0.9598,0.9598]
pvcrec=[0.9547,0.9625,0.9599,0.9599,0.9599,0.9599]
pvcfsc=[0.9525,0.9623,0.9597,0.9574,0.9574,0.9574]
pvcacc=[0.95,0.96,0.96,0.96,0.96]
```

```python
plt.plot(tit,pvcpre)
plt.xlabel('PCA component')
plt.ylabel('Precision')
plt.scatter(400,nopvc[0],c='orange')
plt.text(470, nopvc[0]-0.0006, "No PCA", fontsize=12, color = "b", style = "italic", weight = "light", verticalalignment='center', horizontalalignment='right')
plt.title('PCA precision')
plt.savefig(os.path.join('./visualize/pre.jpg'),
             dpi=500,
             bbox_inches='tight',
             pad_inches=0)
```

```python
plt.plot(tit,pvcrec)
plt.xlabel('PCA component')
plt.ylabel('Recall')
plt.scatter(400,nopvc[1],c='orange')
plt.text(470, nopvc[1]-0.0005, "No PCA", fontsize=12, color = "b", style = "italic", weight = "light", verticalalignment='center', horizontalalignment='right')
plt.title('PCA recall')
plt.savefig(os.path.join('./visualize/recall.jpg'),
             dpi=500,
             bbox_inches='tight',
             pad_inches=0)
```

```python
plt.plot(tit,pvcfsc)
plt.xlabel('PCA component')
plt.ylabel('F1-score')
plt.scatter(400,nopvc[2],c='orange')
plt.text(470, nopvc[2]-0.0005, "No PCA", fontsize=12, color = "b", style = "italic", weight = "light", verticalalignment='center', horizontalalignment='right')
plt.title('PCA f1-score')
plt.savefig(os.path.join('./visualize/f1score.jpg'),
             dpi=500,
             bbox_inches='tight',
             pad_inches=0)
```
